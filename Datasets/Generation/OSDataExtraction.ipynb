```python
from nltk.corpus.reader.bracket_parse import BracketParseCorpusReader
from collections import Counter
import numpy as np
```

```python
problems = []
def helipad_extract(root, files):
  data = BracketParseCorpusReader(root, files)
  sequences = data.tagged_sents()
  clean_seqs = []
  for seq in sequences:
    x = [word for word in seq if (('<' in word[0]) == False) and (('>' in word[0]) == False) and (('*' in word[0]) == False) and (('OSHeliand' in word[0]) == False) and (('0' in word[0]) == False)]
    clean_seqs.append(x)
  sentences, sentence_tags =[], []
  for tagged_sentence in clean_seqs:
    sentence, tags = zip(*tagged_sentence)
    sentences.append(list(sentence))
    sentence_tags.append(list(tags))
  clean_sents, clean_tags = [], []
  for sent in sentences:
    x = [str(word).split('-')[0] if '-' in word else word for word in sent]
    x = ['uuord' if (word == 'N^A^SG') else word for word in x]
    clean_sents.append(x)
  for sent in sentence_tags:
    x = sent
    x = [str(word).split('^')[0] for word in sent]
    x = [str(word).split('+')[1] if (('+' in word) == True) else word for word in x]
    x = [',' if (word == '"') else word for word in x]
    x = ['N' if (word == 'uuord-word') or (word == 'NPR') or (word == 'MAN') else word for word in x]
    x = ['XX' if (word == 'VBG') or (word == 'WADVP') or (word == 'UTP') or (word == 'RN') or (word == 'MG') or (word == 'ALSO') or (word == 'FW') else word for word in x]
    x = ['VB' if (word == 'BAG') or ('BE' in word) else word for word in x]
    x = ['VB' if (word == 'HAG') or ('HV' in word) else word for word in x]
    x = ['VB' if ('AX' in word) else word for word in x]
    x = ['VB' if ('RD' in word) else word for word in x]
    x = ['VB' if ('MD' in word) else word for word in x]
    x = ['VB' if (word == 'VAG') or ('VB' in word) else word for word in x]
    x = ['VB' if ('VN' in word) or ('VG' in word) or (word == 'TO') else word for word in x]
    x = ['ADJ' if ('ADJ' in word) else word for word in x]
    x = ['ADV' if ('ADV' in word) else word for word in x]
    x = ['Q' if (word == 'QR') or (word == 'QS') else word for word in x]
    x = ['PRO$' if (word == 'PROS') else word for word in x]
    x = ['PUNCT' if (word == ",") or (word == '.') or (word == "'") else word for word in x]
    x = ['WH' if (word == 'WPRO') or (word == 'WQ') else word for word in x]
    x = ['PART' if (word == 'RP') else word for word in x]
    clean_tags.append(x)
  for idx, sent in enumerate(clean_sents):
    sent.insert(0, 'start')
    sent.append('stop')
    clean_sents[idx] = sent
  for idx, sent in enumerate(clean_tags):
    sent.insert(0, 'start')
    sent.append('stop')
    clean_tags[idx] = sent
  for idx, sent in enumerate(clean_tags):
    if 'XX' in sent:
      problems.append(idx)
  return clean_sents, clean_tags
```

```python
#download the original file from https://zenodo.org/records/4395040
root = 'helipad'
files = ['heliand.psd']
sents, tags = helipad_extract(root, files)
```

```python
allwords = [word for sent in sents for word in sent]
alltags = [tag for sent in tags for tag in sent]
```
